## Apply Fourier transform on multiple images

### For cleaner fft we do some transformations
```
function [f] = fourier_transform(img)
    f = fft2(img);
    f = fftshift(f);
    f = abs(f);
    f = log(f+1);
    f = mat2gray(f);
    imshowpair(img,f,'montage');
    
end
```

### Img1a

````matlab
img = imread('./Img1a.png');
fourier_transform(img);
````

### Img1b

````matlab
img = imread('./Img1b.png');
fourier_transform(img);
````

### Intresting result
- We can notice that the rectangle shifted but the fourier domain of the image doesn't change 
- Because fourier domain doesnt depend on the location of the wave

### Img2a`

````matlab
img = imread('./Img2a.png');
fourier_transform(img);
````

### Img2b

````matlab
img = imread('./Img2b.png');
fourier_transform(img);
````

- Bluring affects both time and fourier domain

## Removing salt and pepper noise using median filter

````matlab
img = imread('./Img3.png');
for i = 1:3
new_img(:,:,i) = medfilt2(img(:,:,i));
end
imgaussfilt(new_img,1.5);
imshowpair(img,new_img,'montage');
````
